```python
from sympy.physics.mechanics import dynamicsymbols, init_vprinting
init_vprinting()
t = Symbol('t')
theta , phi = dynamicsymbols('theta phi')
dt = diff(theta,t)
dp = diff(phi,t)
dt,dp
```

```python
R = Matrix([[sin(theta)*cos(phi), sin(theta)*sin(phi), cos(theta)],
            [cos(theta)*cos(phi), cos(theta)*sin(phi), -sin(theta)],
            [-sin(phi), cos(phi), 0]])
R
```

```python
dR = diff(R,t)
dR
```

```python
W = simplify(dR * R.T).T
W
```

```python
simplify(R*R.T)
```

```python
simplify(diff(R.inv(),t)*R)
```
