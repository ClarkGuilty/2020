```python
from sympy import *
init_printing()
```

```python
from sympy.physics.mechanics import dynamicsymbols, init_vprinting
init_vprinting()
t = Symbol('t')
theta , phi, psi = dynamicsymbols('theta phi psi')
```

```python
I0, I3 = symbols('I_1 I_3')
phi, theta, psi, I0, I3
```

```python
#Rotación alrededor de Z un ángulo phi.
Rphi = Matrix([[cos(phi), -sin(phi),0], [sin(phi), cos(phi), 0],[0,0,1]])
Rphi
```

```python
#Rotación alrededor de Z un ángulo psi.
Rpsi = Rphi.subs(phi,psi)
Rpsi
```

```python
#Rotación alrededor de X un ángulo theta.
Rtheta = Matrix([[1,0,0], [0, cos(theta), -sin(theta)], [0, sin(theta), cos(theta)]])
Rtheta
```

```python
#Rotación completa en términos de los ángulos de Euler.
R = simplify(Rphi * Rtheta * Rpsi)
R
```

```python
#Momento de inercia en el marco del cuerpo.
I = Matrix([[I0,0,0],[0,I0,0],[0,0,I3]])
I
```

```python
#Momento de inercia en el marco espacial.
Ie = simplify(R * I * R.T)
Ie 
```

```python
#wx en su forma matricial. W = dR/dt * R^T
W = simplify(diff(R,t)*R.T)
W
```

```python
#Vector w.
W[2,1],W[1,0], W[0,2]
```

```python
#Vector w a partir de los componentes de W. Esto es en la base espacial.
w = Matrix([W[2,1],W[0,2], W[1,0]])
w
```

```python
#w en la base del cuerpo.
w_c = simplify(R.T * w)
w_c
```

```python
# w^T I w en la base espacial.
simplify(w.T*Ie*w)
```

```python
# w^T I w en la base fija al cuerpo. 
simplify( w_c.T*I*w_c )
```
